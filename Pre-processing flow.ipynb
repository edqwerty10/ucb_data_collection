```python
def get_coords_from_address(address):
    payload = {
        'address': address, 
        'key': API_KEY
    }
    print(address)
    request = requests.get(GOOGLE_MAPS_URL, params = payload).json()
    results = request['results']

    lat = None
    lng = None
    
    if len(results):    
        answer = results[0]
        lat = answer.get('geometry').get('location').get('lat')
        lng = answer.get('geometry').get('location').get('lng')

    return lat, lng
```

```python
def address_to_coordinate(wrap):
    main_road, start, end = wrap[["Main road", "From", "To"]]
    print(main_road)
    start_lat, start_lng = get_coords_from_address(main_road + " & " + start + ", Fremont CA")
    end_lat, end_lng = get_coords_from_address(main_road + " & " + end + ", Fremont CA")
    return np.array([start_lat, start_lng, end_lat, end_lng])
```
